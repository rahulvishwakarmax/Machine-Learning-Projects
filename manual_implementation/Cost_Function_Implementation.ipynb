**In this lab we will implement the *cost* function for linear regression.**

The cost function measures how well our linear regression model is fitting the training data. In this notebook, we'll visualize how the cost function changes with different parameter values.

```python
import numpy as np
import matplotlib.pyplot as plt
# Set the style for plots
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2]) # (size in 1000s sqft)
y_train = np.array([250, 300, 480, 430, 630, 730]) # (price in $1000s)
```

The equation for cost with one variable is:
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$ 
 
where 
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{2}$$
  
- $f_{w,b}(x^{(i)})$ is our prediction for example $i$ using parameters $w,b$.  
- $(f_{w,b}(x^{(i)}) -y^{(i)})^2$ is the squared difference between the target value and the prediction.   
- These differences are summed over all the $m$ examples and divided by `2m` to produce the cost, $J(w,b)$.  

```python
# Defining the cost function

def compute_cost(x, y, w, b):
    """
    Computes the cost function for linear regression.
    Args:
      x (ndarray (m,)): Data, m examples 
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters  
    Returns
        total_cost (float): The cost of using w,b as the parameters for linear regression
               to fit the data points in x and y
    """
    # nunber of training example
    m = x.shape[0]

    cost_sum = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum = cost_sum + cost
    total_cost = (1 / (2 * m)) * cost_sum

    return total_cost
```

```python
# visualzing the data points
plt.scatter(x_train, y_train, color='blue', marker='x')
plt.title('Training Data: House Prices vs. Size')
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price ($1000s)')
plt.grid(True)
plt.show()
```

```python
# Visualize Cost Function with Varying Weight
# We keep b fixed, and vary w to see how cost changes
w_values = np.linspace(0, 400, 100)  # Range of w values to explore
b = 0  # Fix bias (intercept) at 0
cost_values = [compute_cost(x_train, y_train, w, b) for w in w_values]

plt.plot(w_values, cost_values, color='green')
plt.title('Cost vs. w')
plt.xlabel('w (weight)')
plt.ylabel('Cost')
plt.grid(True)
plt.show()
```
