**Linear Model Implementation**

In this notebook, we will implement and explore the linear model \( f_{w,b}(x) = w \cdot x + b \) for a single variable. 

*Objectives*
- Implement the model
- Visualize predictions with different parameters

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x_train is the input variable(size in 1000 sq feet)
# y_train is the target variable(pricee in 1000s of dollars)
x_train = np.array([1.0,2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0] # m is the no, of training example
print(f"Number of training examples is : {m}")
```

```python
# we can also use len() function
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
i = 0 # change this to 1 to see (x^1, y^1)
x_i = x_train[1]
y_i = y_train[1]
print(f"(x^({1}), y^({1})) = ({x_i}, {y_i})")
```

```python
# plot the data points
plt.scatter(x_train,y_train, marker='x',c='r')
# set the title
plt.title("Housing Prices")
#set the x-axis and y-axis label
plt.ylabel('Price(in 1000s of dollars)')
plt.xlabel('Size (1000s sqft)')
plt.show()
```

```python
w = 100
b = 100
print(f"w : {w}")
print(f"b : {b}")
```

```python
def compute_model_output(x, w,b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

```python
# call the compute_model_output function
tmp_f_wb = compute_model_output(x_train,w,b)
# plot the model prediction
plt.plot(x_train, tmp_f_wb, c = 'b' , label = 'Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker = 'x' , c = 'r', label = 'Actual Values')

# Set the title
plt.title("Housing Prices")

# Set the x_axis and y_axis label
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
